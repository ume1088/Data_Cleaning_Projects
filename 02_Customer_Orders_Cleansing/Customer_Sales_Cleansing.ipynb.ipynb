# Data Cleaning

### Order Date type Fixing

```python
df['Order Date'] = pd.to_datetime(df_before['Order Date'], 
                                   dayfirst=True,
                                   errors='coerce')
```

```python
df.dtypes
```

```python

df['Year']  = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')
```

### Duplicates Handling

```python
df[df.duplicated(keep=False)]
```

# Missing Value Handling

```python
df.isnull().all(axis=1).sum()
```

```python
df=df.dropna(how="all")
```

```python
df.isnull().all(axis=1).sum()
```

### Filling Missing Discount with Zero(No discount)

```python
df['Discount']=df['Discount'].fillna(0)
```

```python
df.isnull().sum()
```

## Email and Phone Number validation

```python
df['Phone Number'].head(20)
```

```python
df['Email'].head(20)
```

```python
# Step 1 — See what prefix looks like
print(df['Phone Number'].str.startswith('Tel:').all())  # are ALL starting with Tel?

# Step 2 — Clean: remove "Tel: " prefix
df['Phone Number'] = df['Phone Number'].str.replace('Tel: ', '', regex=False).str.strip()

# Step 3 — Validate: must be exactly 10 digits
df['Phone Valid'] = df['Phone Number'].str.isdigit() & (df['Phone Number'].str.len() == 10)

# Step 4 — View invalid ones
print(df[df['Phone Valid'] == False][['Phone Number']])

# Step 5 — Check
print(df['Phone Valid'].value_counts())
```

```python
# Clean
df['Email'] = df['Email'].str.lower().str.strip()
df['Email'] = df['Email'].str.replace(' ', '', regex=False)

# Validate
import re
pattern = r"^[\w\.\s']+@[\w]+\.[\w]{2,4}$"
df['Email Valid'] = df['Email'].str.match(pattern)

# Results
print(df['Email Valid'].value_counts(dropna=False))
print(df[df['Email Valid'] == False][['Email']].head(10))
```

### 4. Data Type Issues
- **Order Date** stored as string  
  → Dataset contained mixed date formats (MM/DD/YYYY and DD/MM/YYYY)  
  → Converted using `pd.to_datetime(dayfirst=True, errors='coerce')`

  → Prefix removed, stored as clean string

### 5. Inconsistent Formatting
- **City column** — all values in lowercase  
  → Standardized to Title Case (`jeddah` → `Jeddah`)
